```python
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year
y = parse_year(date)
print(type(y))
y
```

```python
parse_year(date)
```

```python
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)
    return month
parse_month(date)
```

```python
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
```
